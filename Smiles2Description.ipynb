```python
from openai import OpenAI
import pandas as pd
import csv
import response
```

```python
# Load the training data
df_data = pd.read_csv('mol_test.csv')

# Extract SMILES and Target from the training data
smiles_list = list(df_data["SMILES"])[0]
target_list = list(df_data["TARGET"])[0]

client = OpenAI(
    # defaults to os.environ.get("OPENAI_API_KEY")
#     api_key="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",
    api_key="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",
)

# Open the CSV file in append mode with header
with open('output_25.csv', 'a', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)

    # Check if the file is empty
    is_empty = csvfile.tell() == 0

    # Write header if the file is empty
    if is_empty:
        csvwriter.writerow(["Smiles", "Target", "Description"])

    # Iterate through each SMILES and Target
    for SMILES, target in zip(smiles_list, target_list):
        chat_completion = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "user", "content": f"'{SMILES}'包含哪些结构片段"},
            ]
        )

        # Get the description from the API response
        description =  chat_completion.choices[0].message.content

        # Write the result to the CSV file
        csvwriter.writerow([SMILES, target, description])

        print(f"Description for {SMILES} written successfully. Description: {description}")
```
